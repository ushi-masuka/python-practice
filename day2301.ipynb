### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.

```python
# code here
df = pd.read_csv('https://tinyurl.com/2fe6vz4u')
df.head()
```

```python
IData=df[df["Country"]=="Iran"].groupby("Year").agg({"PM2.5":"sum"}).reset_index()
CData=df[df["Country"]=="China"].groupby("Year").agg({"PM2.5":"sum"}).reset_index()
IData
```

```python
plt.plot(IData["Year"],IData["PM2.5"],label="Iran",marker="o")
plt.plot(CData["Year"],CData["PM2.5"],label="China",marker="o")

plt.legend()
plt.title("Year vs. Pol(PM2.5) data")
plt.xlabel("Year")
plt.ylabel("PM2.5 sum(over the year)")
plt.show()
```

### **`Problem-2:`** Draw a histogram of the  column "PM10" of which the y-axis represents the probability (see the documentation how to draw the probability).

```python
plt.hist(df["PM10"],bins=[i for i in range(0,600,10)],density=True,alpha=1,color="blue")
plt.title("probability hist of PM10")
plt.xlabel("PM10 value")
plt.ylabel("probabilty of a PM10 value")
plt.show()
```

### **`Problem-3:`** Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.

```python
plt.scatter(df[df["Country"]=="Poland"]["PM2.5"],df[df["Country"]=="Poland"]["PM10"],marker="+",color="red",label="Poland")
plt.scatter(df[df["Country"]=="Chile"]["PM2.5"],df[df["Country"]=="Chile"]["PM10"],color="cyan",label="Chile")
plt.legend()
plt.xlabel("PM2.5")
plt.ylabel("PM10")
plt.title("PM2.5 vs PM10 in Poland and Chile")
plt.grid()
```

### **`Problem-4:`** Draw a pie chart of top 5 most frequent countries.

```python
fc=df.Country.value_counts().head()
plt.pie(fc.values,labels=fc.index,autopct="%0.1f%%")
plt.title("frequent countries")
plt.show()
```

### **`Problem-5:`** Draw a bar chart which represents the counts of top 5 most frequent countries.
